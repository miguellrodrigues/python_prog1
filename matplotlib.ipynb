```python
# O Matplotlib é uma biblioteca para desenhar gráficos.

# Importando a biblioteca matplotlib
import numpy as np
import matplotlib.pyplot as plt  # Abreviação para plt
```

Plotando linhas

```python
x = np.linspace(-5, 5, 20)  # Cria um array de 11 elementos
y = x ** 2  # Calcula o quadrado de cada elemento

plt.plot(x, y)  # Plota o gráfico
```

```python
plt.plot(x, y, 'o--', color='red')  # Plota o gráfico
```

# Criando figura

```python
plt.figure(figsize=(8,3))  # Cria uma nova figura
plt.plot(x, y, 'o--', color='red', label='f(x)')
plt.xlabel('x')  # Adiciona um rótulo ao eixo x
plt.ylabel('y')  # Adiciona um rótulo ao eixo y
plt.title('Gráfico de x^2')  # Adiciona um título ao gráfico
plt.legend(loc='lower left', fontsize=12)  # Adiciona uma legenda no gráfico
```

```python
# Ajustando os limites de x e y
plt.figure(figsize=(8,3))  # Cria uma nova figura
plt.plot(x, y, 'o--', color='red', label='f(x)')
plt.xlim(-5, 5)  # Define o limite do eixo x
plt.ylim(-5, 5)  # Define o limite do eixo y
plt.xlabel('x')  # Adiciona um rótulo ao eixo x
plt.ylabel('y')  # Adiciona um rótulo ao eixo y
plt.title('Gráfico de x^2')  # Adiciona um título ao gráfico
plt.legend(loc='lower left', fontsize=12)  # Adiciona uma legenda no gráfico
```

```python
# Multiplos gráficos na mesma figura
fig, axes = plt.subplots(3, 2, figsize=(8,8))
ax = axes[0][0]
ax.plot(x, y, 'o--', color='r', lw=0.4, ms=3)
ax.text(0.1, 0.1, 'texto', transform=ax.transAxes)
ax = axes[1][1]
ax.plot(y, x, label='C1')
ax.plot(y/2, x, label='C2')
fig.suptitle('Titulo da figura', y=1.03, fontsize=25)
plt.show()
```

```python
# Salvando figura
# Deve-se salvar a figura antes de usar a funcao plt.show()
plt.savefig('figura.png', dpi=300)
# DPI = Densidade de pixels por polegadas
```
